```python
import numpy as np
from scipy import stats
arr1=np.array([1,2,3,4,5,5])
print(arr1)
mn=np.mean(arr1)
print("The mean of the array is:",mn)
med=np.median(arr1)
print("The median in the array is:",med)
mod=stats.mode(arr1)
print("The mode in the array is:",mod)
```

```python
import numpy as np

data = [1, 2 , 3 , 4 , 5 ]


variance = np.var(data)


std_deviation = np.std(data)

print("Variance (using library function):", variance)
print("Standard Deviation (using library function):", std_deviation)
```

```python
data = [1 , 2,3,6,8,9]


mean = sum(data) / len(data)


squared_diff = [(x - mean) ** 2 for x in data]


variance = sum(squared_diff) / len(data)


std_deviation = variance ** 0.5

print("Variance (without library function):", variance)
print("Standard Deviation (without library function):", std_deviation)
```
